### ðŸŽ¯ Group Training (New Feature)

Train models on specific label groups for hierarchical classification tasks. Currently supported for `chestmnist` dataset.

**ChestMNIST Label Groups:**
- Group 0: Classes [0, 1, 2] (3 classes)
- Group 1: Classes [3, 4, 5, 6, 7, 8] (6 classes)
- Group 2: Classes [9, 10, 11, 12, 13] (5 classes)

#### Train a specific group:

````python
# Train only Group 0 (Classes 0, 1, 2)
!python main.py --model_name 'MedViT_small' --dataset 'chestmnist' --group 0 --epochs 50
````

````python
# Train only Group 1 (Classes 3-8)
!python main.py --model_name 'MedViT_small' --dataset 'chestmnist' --group 1 --epochs 50
````

#### Train all groups automatically using shell script:

````python
# This will train all 3 groups sequentially
!bash train_all_groups.sh
````
